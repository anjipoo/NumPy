```python
import numpy as np
```

## 1. reshape `.reshape(row,col)`
- if possible then reshape kardega

```python
arr1=np.arange(100,200)
arr1
```

```python
arr1=arr1.reshape(10,10)
arr1
```

## 2. slicing 
- if row only: `arr[x]`
- for col: `arr[:,x]` where ":" is for row and uske baad ka is for col

```python
arr1[3] #3rd row
```

```python
arr1[:5] #first 5 rows
```

```python
#print 3rd col
arr1[:,3]
```

```python
arr1[2:3, 4:5] #for particular item
```

## 3. iteration `np.nditer()`
- ismein array pass karo

```python
for i in arr1:
    print(i)
```

```python
for i in np.nditer(arr1):
    print(i)
```
