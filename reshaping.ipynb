## 3. stacking `np.hstack((arrays))` or `np.vstack((arrays))`
- stack karte hai arrays ko
- shapes must match

```python
arr5=np.arange(6).reshape(2,3)
arr6=np.arange(12,18).reshape(2,3)
```

```python
np.hstack((arr5,arr6))
```

```python
np.vstack((arr5,arr6))
```
