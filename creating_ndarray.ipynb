# ways to create arrays in np

## 1. `np.array()`
- list pass karte hai


```python
arr1=np.array([1,2,3,4])
print(arr1)
print(type(arr1))
```

```python
arr2=np.array([[1,2,3],[4,5,6]])
print(arr2)
print(type(arr2))
```

## 2. zeros `np.zeros()` and ones `np.ones()`
- no. of rows and cols passed

```python
arr_zeros=np.zeros((2,3)) #zeros as all elem
print(arr_zeros)

#similaryly for 1's
arr_ones=np.ones((3,2))
print(arr_ones)
```

## 3. identity matrix `np.identity()`
- no. of rows passed (idmatrix is nxn)

```python
#identity matrix
arr_identity=np.identity(5)
print(arr_identity)
```

## 4. range based `np.arange()`
- ismein start, end, steps pass karte hai like normal range() 

```python
#range based
arr6=np.arange(5,10,2) #5 -> begin, 10-> end, 2-> jumps
print(arr6)
```

## 5. linspace `np.linspace()`
- 3 inputs start, end & how many samples to generate
- has more params - 
1. `endpoint=False` which means end wala nahi consider karna 
2. `retstep=True` matlab wo batayega whats the step size

```python
arr7=np.linspace(5,10,10)
print(arr7)

#using retstep
arr7, steps=np.linspace(5,10,10,retstep=True)
print(arr7)
print("steps:",steps)
```

## 6. copy `copy()`
- seedha dupe banata hai

```python
arr8=arr7.copy()
print(arr8)
```

# properties of np arrays
shape, ndim, size, itemsize, dtype, astype()

## 1. shape `.shape`
- output is (rows, cols, no.of arrays)

```python
print(arr1)
print(arr1.shape) #1d array hai
```

```python
arr9=np.array([[[1,2], [3,4]],[[5,6],[7,8]]])
print(arr9)
print(arr9.shape)
```

## 2. ndim `.ndim()`
- dimensions return karta hai

```python
print(arr2)
print(arr2.ndim)
```

```python
print(arr9)
print(arr9.ndim)
```

## 3. size `.size()`
- seedha no. of elems

```python
print(arr9)
print(arr9.size)
```

## 4. data type `.dtype`
- data type batata hai

```python
print(arr_ones)
print(arr_ones.dtype)

print(arr9)
print(arr9.dtype)
```

## 5. itemsize `.itemsize()`
- ye batata hai size of each item
- dtype ke hisaab se

```python
print(arr_zeros)
print(arr_zeros.itemsize)
```

## 6. change type `.astype()`
- ek se dusre mein change karega

```python
print(arr9)
print(arr9.dtype)
arr9=arr9.astype('float')
print(arr9)
print(arr9.dtype)
```
