# broadcasting

```python
import numpy as np
```

```python
a=np.arange(8).reshape(2,4)
b=np.arange(8,16).reshape(2,4)
print(a)
print(b)
```

```python
a+b
```

```python
c=np.arange(9).reshape(3,3)
d=np.arange(3).reshape(1,3)
print(c)
print(d)
```

```python
c+d
```

## ability of numpy to treat arrays of diff shapes during operations
- dissimilar arrays honge to broadcasting hota hai
- chota array is broadcasted into bigger one

```python
print(a.shape)
print(b.shape)
print(a+b)
```

```python
x=np.arange(3).reshape(1,3)
y=np.arange(12).reshape(4,3)
print(x+y)
```

## basically one of the two dims needs to be same OR one has to be 1x1

```python
x=np.arange(3,4).reshape(1)
y=np.arange(12).reshape(4,3)
x+y
```
